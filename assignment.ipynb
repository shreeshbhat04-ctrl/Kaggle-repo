```python
# Import the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Question 1: NumPy Computation

This section demonstrates basic NumPy operations to calculate statistical measures on a one-dimensional array.

```python
# Create a one-dimensional array with at least six numbers
arr = np.array([10, 25, 30, 45, 50, 15, 60])
print(f"The array is: {arr}")

# Calculate and display the maximum value
max_value = np.max(arr)
print(f"Maximum value: {max_value}")

# Calculate and display the minimum value
min_value = np.min(arr)
print(f"Minimum value: {min_value}")

# Calculate and display the mean
mean_value = np.mean(arr)
print(f"Mean: {mean_value:.2f}")

# Calculate and display the standard deviation
std_dev = np.std(arr)
print(f"Standard deviation: {std_dev:.2f}")
```

#### Results for Question 1:
The program successfully created a NumPy array and computed its maximum, minimum, mean, and standard deviation. These statistical measures provide a quick summary of the data distribution within the array.

### Question 2: Data Analysis and Visualization

This section uses Pandas to create and analyze tabular data, and Matplotlib to visualize student marks.

```python
# 1. Create a DataFrame with Student Name and Marks
data = {
    'Student Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace'],
    'Marks': [85, 72, 90, 68, 95, 78, 88]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
display(df)

# 2. Calculate the average marks
average_marks = df['Marks'].mean()
print(f"\nAverage Marks: {average_marks:.2f}")

# 3. Display the students who scored above the average marks
students_above_average = df[df['Marks'] > average_marks]
print("\nStudents who scored above average:")
display(students_above_average)

# 4. Plot a bar chart showing the marks of each student
plt.figure(figsize=(10, 6))
plt.bar(df['Student Name'], df['Marks'], color='skyblue')
plt.xlabel('Student Name')
plt.ylabel('Marks')
plt.title('Student Marks Distribution')
plt.ylim(0, 100) # Assuming marks are out of 100
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.tight_layout()
plt.show()
```

#### Results for Question 2:
The Pandas DataFrame was successfully created, and the average marks were calculated. Students scoring above the average were identified and displayed. The bar chart visually represents each student's marks, making it easy to compare performance.
