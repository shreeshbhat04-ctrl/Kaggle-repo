```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
X_full = pd.read_csv('home-data-for-ml-course/train.csv', index_col='Id')
X_test_full = pd.read_csv('home-data-for-ml-course/test.csv', index_col='Id')
```

```python
X_full.describe()
```

```python
y = X_full.SalePrice
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = X_full[features].copy()
X_test = X_test_full[features].copy()
```

```python
X.head()
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2,
                                                      random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
mymodel= RandomForestRegressor(n_estimators=200, min_samples_split=20, random_state=0)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python

def score_model(model, X_t=X_train, X_v=X_valid, y_t=y_train, y_v=y_valid):
    model.fit(X_t, y_t)
    preds = model.predict(X_v)
    return mean_absolute_error(y_v, preds)
```

```python
mae=score_model(mymodel)
print("Model's MAE: %d" % (mae))
```

```python
mymodel.fit(X, y)
# Generate test predictions
preds_test = mymodel.predict(X_test)
```

```python
output = pd.DataFrame({'Id': X_test.index,
                       'SalePrice': preds_test})
```

```python
output.to_csv('submission2.csv', index=False)
```

```python
import pandas as pd

X_full = pd.read_csv('home-data-for-ml-course/train.csv')
X_test_full = pd.read_csv('home-data-for-ml-course/test.csv')
```

```python
from sklearn.model_selection import train_test_split

X_train1, X_valid1, y_train1, y_valid1 = train_test_split(
    X, y, train_size=0.8, test_size=0.2, random_state=42
)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import HistGradientBoostingRegressor
```

```python
import numpy as np
cat_cols = [c for c in X.columns if X[c].dtype == "object"] 
num_cols = [c for c in X.columns if X[c].dtype != "object"]
preprocessor = ColumnTransformer( transformers=[ ("num", SimpleImputer(strategy="median"), num_cols), ("cat", Pipeline([ ("imputer", SimpleImputer(strategy="most_frequent")), ("onehot", OneHotEncoder( handle_unknown="ignore", sparse_output=False )) ]), cat_cols) ] ) # Model 
model1 = HistGradientBoostingRegressor( learning_rate=0.03, max_depth=10, max_iter=800, min_samples_leaf=20, l2_regularization=0.01, random_state=42 ) # Pipeline 
pipeline = Pipeline([ ("prep", preprocessor), ("model", model1) ]) 
pipeline.fit(X_train1, y_train1)
val_preds = pipeline.predict(X_valid1) 
mae = mean_absolute_error(y_valid1, val_preds)
print(f"Validation MAE: {mae:.2f}") 
pipeline.fit(X, y) 
test_preds = pipeline.predict(X_test_full) 
output3 = pd.DataFrame({ "Id": X_test_full["Id"], "SalePrice": test_preds })
```

```python
output3.describe()
```

```python
print(f"Validation MAE: {mae:.2f}") 
```

```python
output3.to_csv('submission2.2.csv', index=False)
```
