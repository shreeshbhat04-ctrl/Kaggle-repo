```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
X_full = pd.read_csv('home-data-for-ml-course/train.csv', index_col='Id')
X_test_full = pd.read_csv('home-data-for-ml-course/test.csv', index_col='Id')
```

```python
X_full.describe()
```

```python
y = X_full.SalePrice
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = X_full[features].copy()
X_test = X_test_full[features].copy()
```

```python
X.head()
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2,
                                                      random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
mymodel= RandomForestRegressor(n_estimators=200, min_samples_split=20, random_state=0)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python

def score_model(model, X_t=X_train, X_v=X_valid, y_t=y_train, y_v=y_valid):
    model.fit(X_t, y_t)
    preds = model.predict(X_v)
    return mean_absolute_error(y_v, preds)
```

```python
mae=score_model(mymodel)
print("Model's MAE: %d" % (mae))
```

```python
mymodel.fit(X, y)
# Generate test predictions
preds_test = mymodel.predict(X_test)
```

```python
output = pd.DataFrame({'Id': X_test.index,
                       'SalePrice': preds_test})
```

```python
output.to_csv('submission2.csv', index=False)
```
